# SIR Model

### Define initial conditions at the beginning of an outbreak

```python
# Size of population 
n = 50000000

# Fraction of population infected
i_beginning = 1 / n

# Fraction of population who have recovered or died
r_beginning = 0

# Fraction of population who are susceptible
s_beginning = 1 - i_beginning - r_beginning

# Vector containing the initial conditions
y_beginning = s_beginning, i_beginning, r_beginning
```

### Define R0, beta, and gamma

```python
# Define R0
R0 = 3

# Define gamma (1 / infectious time (days))
gamma = 1 / 2.09

# Define beta
beta = R0 * gamma

print("Gamma is: ", round(gamma, 2), " and beta is: ", round(beta, 2))
```

```python
# length of time to model, in days
t = np.linspace(0, 35, 36)

# function to calculate ds/dt, di/dt, and dr/dt
def SIRmodel(y, t, beta, gamma):
    s, i, r = y
    dsdt = -beta * s * i
    didt = beta * s * i - gamma * i
    drdt = gamma * i
    return dsdt, didt, drdt

# solve the differential equations
sol = odeint(SIRmodel, y_beginning, t, args=(beta, gamma))

s, i, r = sol.T
```

```python
# Plot the data
fig = plt.figure(facecolor='w')
ax = fig.add_subplot(111, axisbelow=True)
ax.plot(t, s, 'r', alpha=0.5, lw=2, label='Susceptible')
ax.plot(t, i, 'g', alpha=0.5, lw=2, label='Infected')
ax.plot(t, r, 'b', alpha=0.5, lw=2, label='Removed')
ax.set_xlabel('Time (days)')
ax.set_ylabel('Fraction of population')
ax.set_ylim(-0.1,1)

legend = ax.legend(loc = "upper left")
legend.get_frame().set_linewidth(0.5)


# remove border from plot
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```

# SEIR Model Demonstration

### Define initial conditions at the beginning of an outbreak

```python
# Size of population 
n = 50000000

# Fraction of population exposed
e_beginning = 1/n

# Fraction of population infected
i_beginning = 0

# Fraction of population who have recovered or died
r_beginning = 0

# Fraction of population who are susceptible
s_beginning = 1 - e_beginning - i_beginning - r_beginning

# Vector containing the initial conditions
y_beginning = s_beginning, e_beginning, i_beginning, r_beginning
```

### Define R0, beta, gamma, and delta

```python
# Define R0
R0 = 3

# Define gamma (1 / infectious time (days))
gamma = 1 / 2.09

# Define beta
beta = R0 * gamma

# Define delta (1 / incubation time (days))
delta = 1 / 4.58

print("Gamma is: ", round(gamma, 2), ", beta is: ", round(beta, 2), ", and delta is: ", round(delta,2))
```

```python
# length of time to model, in days
t = np.linspace(0, 130, 131)

# function to calculate ds/dt, de/dy, di/dt, and dr/dt
def SEIRmodel(y, t, beta, gamma, delta):
    s, e, i, r = y
    dsdt = -beta * s * i
    dedt = beta * s * i - delta * e
    didt = delta * e - gamma * i
    drdt = gamma * i
    return dsdt, dedt, didt, drdt

# solve the differential equations
sol = odeint(SEIRmodel, y_beginning, t, args=(beta, gamma, delta))

s, e, i, r = sol.T
```

```python
# Plot data
fig = plt.figure(facecolor='w')
ax = fig.add_subplot(111, axisbelow=True)
ax.plot(t, s, 'r', alpha=0.5, lw=2, label='Susceptible')
ax.plot(t, e, 'y', alpha=0.5, lw=3, label='Exposed')
ax.plot(t, i, 'g', alpha=0.5, lw=2, label='Infected')
ax.plot(t, r, 'b', alpha=0.5, lw=2, label='Removed')
ax.set_xlabel('Time (days)')
ax.set_ylabel('Fraction of population')
ax.set_ylim(-0.1,1)

legend = ax.legend(loc = "upper left")
legend.get_frame().set_linewidth(0.5)

# remove border from plot
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```

# SIR Model Simulation with Erdos-Renyi graph

```python
beta = 1.44

population = 1000

# Generate graph
g = nx.erdos_renyi_graph(population, 0.1)

# Initialize all nodes to Susceptible
for node in g.nodes():
    g.nodes[node]['compartment'] = "Susceptible"
    g.nodes[node]['day of infection'] = 0

# Randomly select the first node to infect 
firstInfected = randint(0,population)  
g.nodes[firstInfected]['compartment'] = "Infected"
g.nodes[firstInfected]['day of infection'] = 1

print("The first node is: ", firstInfected)
```

```python
import collections
degree_sequence = sorted([d for n, d in g.degree()], reverse=True)
degreeCount = collections.Counter(degree_sequence)
deg, cnt = zip(*degreeCount.items())

fig, ax = plt.subplots()
plt.bar(deg, cnt, width=0.80, color="b")
```

```python
from random import sample

dayCounterList = []
numberOfSusceptibleList = []
numberOfInfectedList = []
numberOfRemovedList = []

numberOfDaysToSimulate = 35

dayCounter = 0

while dayCounter < numberOfDaysToSimulate:
    dayCounterList.append(dayCounter)
    print("daycounter is: ", dayCounter)
    infectedNodes = []

    for u in g.nodes():
        if g.nodes[u]['compartment'] == "Infected":
            infectedNodes.append(u)

    print("there are this many infected nodes: ", len(infectedNodes), "\n")
        
    for i in range(0, len(infectedNodes)):
        currentInfectedNode = infectedNodes[i]
        if g.nodes[currentInfectedNode]['day of infection'] == 1 or g.nodes[currentInfectedNode]['day of infection'] == 2:
            # Who are the neighbors of this infected node?
            neighborsOfInfected = g.neighbors(infectedNodes[i])
            listOfNeighbors = list(neighborsOfInfected)
            numberOfNeighbors = len(listOfNeighbors)
            
            # Who are the SUSCEPTIBLE neighbors of this infected node?
            susceptibleNeighbors = []
            for j in range(0,len(listOfNeighbors)):
                if g.nodes[listOfNeighbors[j]]['compartment'] == "Susceptible":
                    susceptibleNeighbors.append(listOfNeighbors[j])
            # How many susceptible neighbors are there?
            numberOfSusceptibleNeighbors = len(susceptibleNeighbors)    
            
            # How many of this node's susceptible neighbors can be infected?
            # Based upon each infected person can infect beta*S/n others
            numberInfectedNeighbors = int(round((beta*(numberOfSusceptibleNeighbors/numberOfNeighbors)),0))
            # What is their node number? Choose x out of all the neighbors
            sequence = [i for i in range(0,numberOfNeighbors)]
            subset = sample(sequence, numberInfectedNeighbors)
            
            if len(subset) > 0:
                # Infect the neighbors
                for k in range(0, len(subset)):
                    indexPosition = subset[k]
                    neighborToInfect = listOfNeighbors[indexPosition]
                    g.nodes[neighborToInfect]['compartment'] = "Infected"
                    g.nodes[neighborToInfect]['day of infection'] = 1
                
            # Increment day of infection for the infected node that caused this
            g.nodes[currentInfectedNode]['day of infection'] = g.nodes[currentInfectedNode]['day of infection'] + 1
            
        elif g.nodes[currentInfectedNode]['day of infection'] == 3:
            g.nodes[currentInfectedNode]['compartment'] = "Removed"
            continue
    
    susceptibleCounter = 0

    for u in g.nodes():
        if g.nodes[u]['compartment'] == "Susceptible":
            susceptibleCounter = susceptibleCounter + 1
    
    numberOfSusceptibleList.append(susceptibleCounter/population)
    print("number of susceptible: ", susceptibleCounter)
    
    
    infectedCounter = 0

    for u in g.nodes():
        if g.nodes[u]['compartment'] == "Infected":
            infectedCounter = infectedCounter + 1
    
    numberOfInfectedList.append(infectedCounter/population)
    print("number of infected: ", infectedCounter)
    
    
    removedCounter = 0

    for u in g.nodes():
        if g.nodes[u]['compartment'] == "Removed":
            removedCounter = removedCounter + 1

    numberOfRemovedList.append(removedCounter/population)
    print("number of removed: ", removedCounter)
        
    dayCounter = dayCounter + 1
    
```

```python
#plt.plot([1,2,3,4,12,15],'g*', [1,4,9,16], 'ro')
plt.plot(numberOfSusceptibleList, 'r', numberOfInfectedList, 'g', numberOfRemovedList, 'b')
plt.show()
```
